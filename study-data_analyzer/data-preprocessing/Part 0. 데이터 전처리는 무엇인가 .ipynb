# Part 0. 데이터 전처리는 무엇인가 (큰 그림부터 잡기)

## 0.1) 데이터 분석 전체 흐름의 정확한 이해
데이터 분석은 단순히 수집 → 전처리 → 분석 같은 기술 순서가 아니다. <br>
질문을 만들고, 확인하고, 다듬고, 전달하고, 결정하는 행동의 연속이다.

1. 데이터 수집 (자료 확보)
    -파일, DB, API 등에서 원천 데이터를 가져온다
    - 이 단계의 목적은 “쓸 수 있는 재료를 확보하는 것”이다
    
2. 탐색 (EDA: 상황 파악)
    - 분포를 확인한다
    - 값이 이상한지 살펴본다
    - 변수 간 관계를 가볍게 확인한다
    - 이 과정에서 질문과 가설이 만들어진다

3. 전처리 (분석 가능 상태로 만들기)
    - 형 변환
    - 결측치/이상치 처리
    - 파생 변수 생성
4. 시각화 (확인 + 전달)
    - 패턴을 명확하게 드러낸다
    - 결과를 설명 가능한 형태로 만든다

5. 모델링 (필요한 경우)
    - 예측, 분류, 추천 등
    - 전처리와 시각화가 잘 되어 있을수록 의미 있는 결과가 나온다

6. 결과 해석 / 의사결정
    - 분석 결과를 해석한다
    - 실제 행동으로 이어질 결정을 내린다

이 순서는 권장 사항이 아니라 필수 흐름이다.

## 0.2) 전처리는 "준비 단계"가 아닌 "전제 조건"이다
전처리는 데이터가 계산·비교·집계·시각화가 가능하도록 구조와 타입을 맞추는 작업이다.

### 전처리를 안 하면 발생하는 문제
1. 계산이 불가능해진다.
    - 숫자로 계산해야 할 값이 문자열이면 곱셈, 합계, 평균이 아예 동작하지 않는다.
2. 결과의 왜곡
    - 같은 의미의 값이 다른 값으로 인식되면, 집계 결과가 쪼개진다.
3. 시각화/모델 오류
    - 그래프가 안 그려지거나 모델 학습 단계에서 오류가 난다.

기억하자. 전처리는 분석을 돕는 보조 작업이 아니라 분석이 성립하기 위한 전제 조건이다.

## 0.3 전처리는 기술이 아니라 사고 과정
전처리는 단순히 astype, fillna, dropna 같은 함수를 외워서 쓰는 작업이 아니다. <br>
전처리의 출발점은 항상 "이 데이터로 어떤 질문에 답하려는가?" 이다.

같은 컬럼이여도
- 매출을 합계로 볼지
- 평균으로 볼지
- 날짜별 추이를 볼지

필요한 dtype, 결측치 처리 방식, 파생 피처가 전부 달라진다.<br>
전처리는 기술이 아니라 질문에 맞게 데이터를 조정하는 사고 과정이다.

## 0.4 전처리 사고 순서 요약
전처리는 감으로 하지 않고, 항상 같은 질문 순서로 진행한다.

1. 구조확인
2. dtype 확인
3. 값의 정리 (Cleaning)
4. 분석 가능한 상태로 만들기

### 1. 구조 확인
이 데이터가 다음과 같은지 확인한다.
- 몇 행, 몇 열인지
- 어떤 컬럼들로 구성되어 있는지
- 테이블 형태가 맞는지

### 2. dtype 확인
각 컬럼이
- 숫자인지
- 문자열인지
- 날짜인지

를 확인한다.

이 단계에서 **“금 이 상태로 계산이 가능한가?**를 본다.

### 3. 값의 정리 (Cleaning)
dtype가 맞지 않거나, 값의 표기가 섞여 있거나, 결측치가 있으면 이 단계에서 정리한다.
- 문자열 정리
- 숫자 파싱
- 결측치 처리
- 범주값 통일

전처리의 대부분은 이 단계에서 발생한다.

### 4. 분석 가능한 상태로 만들기
마지막으로 
- 필요한 컬럼만 남기고 
- 파생 피처를 만들고
- 바로 groupby / 시각화가 가능한지 

를 점검한다.

이 상태가 되면 "이제 분석을 시작해도 되는 데이터"가 된다.
