3. 머신러닝의 종류

3-1 지도학습이란?

- 패턴을 학습하여 새로운 데이터에 대해 예측을 수행하는 방식

3-2 비지도학습이란?

- 정답이 없는 데이터에서 패턴을 찾아내는 방식

3-3 군집화 - 패턴을 보고 유사한 데이터끼리 그룹화하는 기법

- 고객 분류
- 영화 추천
- 음악 추천

강화 학습이란?
- 행동과 보상을 통해 최적의 행동을 학습하는 방식
- 예시: 게임 AI, 자율 주행 자동차

실무에선 파이프라인으로 간다

1. 전처리: 결측치 처리, 인코딩, 스케일링(표준화)
2. 모델 구축: 학습/예측
3. 분석 및 활용: 점수/지표

이걸 한번에 묶는 도구가 sklearn.pipeline.Pipeline.
- 전처리와 모델을 한 덩어리로 묶어 실수를 줄임
- 교차검증/

지도학습인 경우 train/test 분리가 필수적이다.

과적합 vs 과소적합
- 과적합: 훈련 데이터에 너무 맞춰져 새로운 데이터에 일반화가 안되는 상태
- 과소적합: 모델이 너무 단순하여 데이터의 패턴을 제대로 학습하지 못한 상태

요약

1. 데이터를 모으고 보고 패턴을 찾는다.
2. 머신러닝 = h(x) = wx + b에서 최적의 w와 b를 찾는다.
3. 손실함수 = "얼마나 틀렸나" -> 를 줄이는게 학습
4. 경사하강법 = 기울기 따라 한걸음씩 내려각
5. 회귀는 수치 예측(MSE/MAE)/분류(LogLoss)는 확률 기반
6. 지도학습 vs 비지도학습 vs 강화학습
7. sklearn.pipeline.Pipeline으로 시행
