# 통계 2차시: 확률과 분포

# 통계 이론 1회차 복습
통계학을 배우는 이유?<br>
통계란 데이터를 요약하고, 비교하고, 의심하고, 판단하는 방법이다.
> 데이터 분석에서 통계는 선택이 아닌 기본기이다.

데이터의 유형
- 수치형 데이터
    - 이산형: 셀 수 있는 값
    - 연속형: 측정을 통해 얻는 값
- 범주형 데이터
    - 명목형: 순서가 없음
    - 순서형: 순서는 있지만 간격의 의미는 없음
    - 이진형: 2가지의 값만 가진다.

기술통계 <br>
수집한 데이터를 정리하고 요약하고 설명(해석과 표현)하는 데 목적이 있다.
>데이터를 보기 쉽게 정리하고 현 상태를 이해하는 데 집중한다.

- 평균: 
    - 산술평균: 모든 값을 더해 데이터 개수로 나눈 평균
    - 절사평균: 값을 정렬한 뒤 양쪽 끝의 일부 값을 제거하고 계산한 평균
    - 가중평균: 각 값에 중요도를 반영해 계산한 평균

- 중앙값: 데이터를 정렬했을 때 가운데에 위치한 값 (개수가 짝수인 경우 가운데 두 값의 평균을 사용)

- 편차: 각 데이터가 평균에서 얼마나 떨어져 있는지를 나타낸 값 
    - 편차 = 각 값 - 평균
    - 그대로 합하면 +, −가 서로 상쇄


- 분산: 편차의 상쇄 문제를 해결하기 위해 편차를 제곱해서 평균낸 값 
    - 분산 = (편차^2들의 평균)
    - 단위가 제곱이라 해석이 직관적이지 않음


- 표준편차: 분산에 제곱근을 취한 값, 사람이 해석하기 쉽게 만든 변이 지표 
    - "보통 평균에서 이 정도 벗어난다"로 해석
    - 표준편차 = √분산

데이터 분포와 탐색
- 백분위수와 박스플롯
- 도수분포표와 히스토그램
- 밀도추정과 밀도그림, 바이올린 도표
- 이진데이터와 범주 데이터 탐색 (막대플롯)

상관관계(산점도 4~5차시에 함 더함)
- 데이터들끼리 서로 관련이 있는지 확인하는 방법
    - 숫자로는 상관계수, 그림으로는 산점도를 사용한다.

- 상관계수
    - +1에 가까울수록: 함께 증가
    - -1에 가까울수록: 하나가 증가하면 다른 하나는 감소
    - 0에 가까울수록: 뚜렷한 관계 없음

추론통계
- 표본을 이용해 모집단의 특성을 추정하는 통계 방법
> 일부 데이터(표본)를 가지고 전체 데이터(모집단)를 판단한다.

# 확률과 확률변수
확률은 불확실한 세상을 이해하는 언어
>우리는 확률을 통해 불확실성 속에서 더 나은 결정을 내리는 방법을 배울것이다.

## 확률이란? — “미래 결과가 얼마나 가능성 있는가”
확률은 불확실한 세상을 이해하는 언어
>우리는 확률을 통해 불확실성 속에서 더 나은 결정을 내리는 방법을 배울것이다.

확률 P(A)
- 0 ≤ P(A) ≤ 1
- 불가능하면 0, 반드시 일어나면 1
- 0과 1 사이 == 사건 A가 일어날 확률 = P(A)
    - P(A) = (사건 A에 해당하는 경우 수) / (전체 경우 수)

구슬뽑기:<br>
구슬 5개중 빨강색 4개 흰색 1개<br>
p(빵강) = 4/5 = 0.8, P(흰색) = 1/5 = 0.2<br>

>확률은 사건이 일어날 가능성의 크기를 숫자로 표현하는 것이다.

### 통계적 실험(Experiment)<br>
우연성이 있는 관측/시도 자체를 말한다.
- ex 동전 던지기, 주사위 굴리기, 고객이 구매했는지 관측하기

비슷한 사건이 반복될 때 여러가지 가능한 결과를 알 수 있지만 정확히 무슨 결과가 발생할 지는 모를 떄, <br>
이러한 특성을 갖는 문제가 바로 통계학의 연구 및 응용 대상이다.

- 결과(Sample Point): 실제로 관측된 하나의 경우

- 표본공간(Sample Space): 가능한 모든 결과(표본점)를 모아둔 집합이다. {}로 표현
    - 이산형 표본공간 == 유한 또는 셀 수 있는 무한
        - 공장에서 제품을 반복 생상할때 정삼품 또는 불량품의 확률
    - 연속횽 표본공간 == 셀 수 없는 무한
        - 물건이 30분 이내로 배달올 확률
    
- 사건(event): 표폰공간의 한 부분집합/그 결과들이 모여있는 집합


생활속의 비유
- 구슬 뽑기: 사건 = 빨간색 구슬 뽑기 / 표본공간 = {빨강, 흰색}<br>
- 가위바위보: 사건 = 가위 / 표본공간 = {가위,바위,보}<br>
- 고객 행동: 사건 = 구매 / 표본공간 = {구메, 미구매}

사건 == 실제 선택한 하나의 결과<br>
표본공간 == 가능한 모든 결과들의 경우의 수

### 대수의 법칙
시행 횟수가 많아질수록, 실제 결과의 평균은 이론적 기댓값에 가까워진다.

예시
- 동전 앞면 확률 = 0.5
- 10번 던지면 → 앞면 7번 나올 수도 있음
- 10,000번 던지면 → 앞면 비율이 0.5에 매우 가까워짐

> 즉 적게 하면 → 운 영향 큼, 많이 하면 → 확률 이론값에 수렴함


시행 결과 평균 → 표본평균<br>
시행을 무한히 많이 하면 표본평균이 기댓값에 가까워짐<br>

## 확률변수란? — “결과를 숫자로 바꾼 것”
사건/결과를 숫자로 치환해 계산 가능하게 만든 변수다.
- 예시: 빨강 4, 흰색 1
    - X= {빨강, 흰색}
    - P(X=빨강) = 0.8
    - P(X=흰색) = 0.2

> 확룰변수는 불확실성을 값으로 다루는 도구로, 표본공간의 각 원소에 하나의 실수값을 대응시키는 함수

이산형 확률변수 
- 값이 0,1,2,…처럼 셀 수 있다
    - ex) 주사위 눈, 불량 개수

연속형 확률변수
- 값이 구간 전체에 걸쳐 연속적으로 존재한다
    - ex) 키, 몸무게, 시간

예시로 확인해보기

예시 1) 주사위(확률 + 사건)
- S = {1,2,3,4,5,6}
- 사건 A: "짝수" = {2,4,6}
- P(A) = 3/6 = 0.5 

예시 2) 동전(확률변수 정의)
동전 결과를 숫자로 바꿔서 확률변수 X 정의
- 앞면이면 X=1
- 뒷면이면 X=0
- 이때 앞면이 나올 확률은 X=1로 표현

# 확률분포 — “값이 나올 확률 구조”
각 값이 나올 확률을 표나 함수로 표현 <br>
확률변수 X가 있을 때 "X가 가질 수 있는 값 + 그 값이 나올 확률"

- 확률변수 = 결과를 숫자로 바꾸는 장치
- 확률분포 = 그 숫자가 어떻게 퍼져있는지 설명

## 이산 확률분포 : 막대 그래프
값의 개수를 셀 수 있으며, <br>
확률을 각각 더하면 = 1

주사위 예시 <br>
P(X=1)=1/6 <br>
P(X=2)=1/6 <br>
P(X=3)=1/6 <br>
P(X=4)=1/6 <br>
P(X=5)=1/6 <br>
P(X=6)=1/6 <br>
→ 모든 확률 합 = 1

- 이항분포 → 성공/실패 반복
- 포아송 → 특정 시간 발생 횟수

## 기댓값 E(x) 
확률분포가 있으면 "평균적으로 얼마 나오냐" 계산 가능하다.<br>
= 확률적으로 기대되는 평균값,  오랫동안 반복했을때 수렴하는 값

- 기대값은 한번의 결과가 아닌 많은 시도(대수의 법칙/표본 크기)의 평균이다.

이산형
- E(X) = Σ xP(x)

연속형
- E(X) = ∫ x f(x) dx

예시 1 — 이산 확률분포 (주사위)
- X = 주사위 눈

- 가능값<br>
1,2,3,4,5,6

- 확률<br>
각각 1/6

- 기대값<br>
E(X) = 3.5

→ 장기적으로 평균 3.5 근처 나온다

예시 2 — 연속 확률분포 (키)

- X = 키

- 가능값<br>
150 ~ 190 사이 무한

- 특정값<br>
P(X=170) = 0

- 구간 <br>
P(165~175) = 의미 있음

## 기억하기
확률변수 = 결과 숫자화<br>
확률분포 = 그 숫자가 나오는 확률 구조

- 분산과 표준편차가 클수록 데이터는 불안정하다
    - 분산과 표준편차는 값들이 얼마나 평균에서 떨어진 정도를 확인하기 때문

## 이론적 확률분포 & 파라미터
사람마다 지문이 다르다, 이는 데이터도 마찬가지로
- 확률분포라는 지문을 보면 집단의 특징이 나타난다.
- 모양을 결정하는 숫자가 존재
- 정규분포는 평균(μ)과 표준편차(σ) 두 숫자로 정의

정리:<br>
모잡단을 N(μ,σ²)로 표현하면, 그 집잔의 성질을 절반은 파악했다 할수 있다.
- N(평균, 분산)


파라미터 <br>
평균μ - 분포의 중심 위치 (대표값): 평균이 올라가면 오른쪽으로 이동 <br>
표준편차σ - 분포의 흩어짐 (안정성): 더 넓고 낮아짐 

> 평균 μ는 분포의 위치를, σ의 모양을 결정한다

# 정규분포 — “현실 데이터의 기본 모양”
종 모양의 대칭 분포로, 대부분의 데이터가 평균 주위에 몰려 있는 분포를 의미
- 평균을 중심으로 좌우 대칭이며, 평균에서 멀어질수록 데이터의 빈도가 감소한다.
- 대부분의 데이터가 평균 주변에 몰려 있으며, 평균에서 멀어질수록 빈도가 줄어든다.

평균 = 중심<br>
표준편차 = 퍼짐 정도

종 모양의 연속함수
- 평균에 대해 서로 대항(좌우 대칭), 따라서 평균의 왼쪽, 오른쪽의 확률은 0.5
- 평균이나 표준편차의 값에 따라 정규분호는 무한히 많이 있음
- 정규확률변수는 평균 주위에 대부분의 값을 가지고, 평균에서 좌우로 표준편차의 3배 이상 떨어진 값은 없음

평균μ - 분포의 중심 위치 (대표값): 평균이 올라가면 오른쪽으로 이동 <br>
표준편차σ - 분포의 흩어짐 (안정성): 더 넓고 낮아짐 

표준정규분포
- 평균 = 0 표준편차 = 1 N(0,1)
- 모든 정규분포의 기준점
- 표준화(z): 모든 데이터를 N(0,1) 눈금으로 전환


### 정리해서 보기
- 정규분포는 세상에서 제일 흔한 분포
- 평균 μ == 위치
- 표준편차σ == 폭


68-95-99 법칙 == 범위를 보면 분포가 보인다.<br>
흔한값과 드문값을 빠르게 가르는 직관적인 기준
- 평균 ±1σ → 68%
- 평균 ±2σ → 95%
- 평균 ±3σ → 99.7%

대부분의 데이터는 평균 근처에 몰리고, 멀리 벗어날수록 드물다.

# Z-score 표준화(z) — “다른 데이터랑 비교 가능하게 만드는 값”
어떤 값이 평균에서 얼마나 떨어져 있는지를, 데이터의 퍼진 정도(표준편차)를 기준으로 나타낸 값이다.<br>
데이터마다 평균과 퍼진 정도가 다르기 때문에, Z-score를 사용하면 데이터를 평균 0, 표준편차 1 기준으로 바꿔서 서로 다른 데이터도 같은 기준에서 비교할 수 있다.

Z = (값 - 평균) / 표준편차<br>
Z = (x-μ)/σ
- 해석: 평균에서 표준편차 몇 배 떨어졌는지 나타내는 숫자<br>
→ 평균에서 몇 표준편차 떨어졌는지

- 평균, 표준편차와 정규분포의 관계
- 정규, 비정규 정규분포는 강력하나 모든 데이터에 항상 맞지 않음 (성격에 따라 구분 가능해야함)
- 정규 근사 (사용 가능)
- 정규 경계 - 오른쪽 꼬리 긴 분포, 해비테이르 이항분포 (주의하며 사용)
- 현실 데이터는 정규보단 꼬리가 긴 경우가 많다
