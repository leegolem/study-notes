- 이는 느낌이 아닌 데이터로 확인해봐야한다. -> 상관관계를 파악

    - 상관계수 (r)로 숫자로 요약하기

## 두 수치형 변수 사이의 관계를 밝히다
현실 데이터는 변수가 쌍으로 존재하는 경우가 많다.<br>
단일변수 평균/분산뿐만 아닌 관계도 함께 봐야함

- 예시

관계 보기란?

2-2) 산점도

- 상관 분석의 출발점: 상관계수(r)

- 상관계수 (r)만 보고 판단하면...
    - 비선형 관계
    - 이상치 하나가 만든 가짜 상관
    - 서로 다른 두 패턴의 혼합<br>를 파악하기 힘듬

- 상관계수의 범위: -1 <=ㄱ <= +1

>숫자는 왜곡 될순 있으나 그림은 직관적으로 관계를 표현한다.

두변수를 동시에 그릴때 가장 기본 도구
- 방향, 강도, 모양

각 데이터 한 점이(x,y) 좌표위에 찍힘
- 우상향 -> 양의 상관
- 우하양 -> 음ㅅ의 상관
- 제멋대로 -> 상관계수가 약함

실무로 연결: 광고비 와 매출<br>
상관계수가 r = 0.6 -> 산점도를 보니 이상치 때문에 과장된 값일 가능성도 있음

2-3) 상관계수<br>
산점도는 관계를 눈으로 확인함으로 보는 사람들 마다 해석이 다를 가능성이 있음<br>
보는 사람마다 다르게 해석하기 때문에 산점도의 패턴을 숫자 하나로 요약하는 숫자, 상관계수가 필요하다.

- 상관계수는 방향(+/-)과 강도를 표현함
- 상관계수: r (-1 ~ +1)
    - 두 수치형 변수가 얼마나 직선적으로 같이 움직여지는지 숫자로 요약한 값

- 예시로 보기: 
    - 공부시간 과 점수 == 0.94 == 양의 상관 (같이 증가)
    - 나이 와 게임시간 == -0.6 == 음의 상관 (한쪽이 증가하면 다른쪽이 감소)
    - 주민등록번호 와 점수 == 0 == 관계를 파악하기 어려움

2-4) 회귀 (가볍게 짚고 넘어가기)

상관은 두 변수가 "같이 움직인다"란 사실만 요약한다.

상관이 알려줄수 없는 영향력과 크기를 알고 싶다면? -> 회귀를 사용함

- 상관: 두 변수를 동등하게 다룸
- 회귀: 한 변수를 설명변수(x), 다른변수를 반응변수(y)로 두고 방향성을 가지어
    - x가 변할때 y는 어떻게 변하는가?

- 상관: 공부시간과 점수가 함께 늘어난다. (같이 움직임)
- 회귀: 공부시간이 1시간 늘 때 점수가 평균 5점 오른다 (영향력의 수치화)

예시로보기: 공부시간 과 집중력
- 0시간 -> 집중력 낮음
- 5시간 -> 집중력 최고점
- 12시간 -> 과로로 집중력 하락

- 곡선 모양이 위로 볼록, 이 관계에선 상관계수가 0에 가까울수 있으나 실제론 강한 비선형 관계

실무예시로 보기: 마케팅 광고비와 매출은 관련이 있음
- 광고비 100만원을 늘리면 매출이 평균 얼마 늘어난다! 같은 구체적 수치가 필요함
- 이때 회귀분석이 강한 도구가 되어줌

2-5) 피어슨 상관계수

가장널리 쓰이는 상관계수: 피어슨 상관계수
- 값의 범위 -1~+1
- 두변수의 선형 관계 강도를 표현
- r = +1: 강한 양의 상관
- r = -1: 강한 음의 상관
- r = 0: 선형 관계 x

Cov(X,Y) / (σX, σY)

- 공분산: 두 변수가 얼마나 같이 움직이나?
- σX, σY: 각 변수의 표준편차
- 공분산의 단위에 따라 크기가 달라짐 -> 표준편차로 나눠 단위를 없애고 -1~+1로 정규화

2-6) 해석 기준

|r| < 0.3 -> 약한 상관 관계<br>
0.3 <= |r| < 0.7 중간 정도<br>
|r|>= 0.6 강한 상관<br>

주의사항

- r이 크다고해 좋다/의미가 반드시 있는것이 아님
- 맥락과 도메인에 따라 해석이 달리짐

보고서 작성시
단순 r = 0.45 -> 중간정도로 라 끝내지 않고

r=0.45로 중간정도의 상관이 관찰되었으며, 이는 고객 행 데이터를 다룰때에 의미 있는 수준이다.

2-7) 이상치의 영향

상관계수는 평균과 분산 기반으로 이상치에 큰 영향을 받는다.

- 숫자만 보고 보고서 작성 금지

# 실험 환경 구축 & 실험단위


실험 환경이란?
- 기능을 노출 할수 있어야함
- 지표를 수집할수 있어야함
- 결과를 비교할 수 있어야함

실험 단위란?
- 실험에 참여하는 최소 단위. 결과를 비교하는 수준
    - 사용자
    - 웹페이지
    - 앱기능
    - 도시/국가 등 거주구역
    - 회사/조직

웹페이지 vs 사용자
- 페이지 단위: 새로고침마다 다른안
- 사용자 단위: 항상 같은 안

좋은 실험 단위 조건
- 가설에 포함된 지표가 측정 가능해야 함
- 실험 단위 간 간선이 없어야 한다
- 실섬의 검정력을 극대화해야함
    - 표본수 충분
    - 간섭의 최소
    - 지표 즉정 가능
- 실험 단위가 정해진 뒤 무작위 할당

예시: 아마존 프라임 요금
- 사용자 단위 실험
- 정보 공유로 간섭 발생

왜 무작위성이여야 하나?<br>
교란 변수 제거, 집단 간 평군적 동질성을 가지기 위해

실험 설계에서 시작
- 가설을 실험으로 번역
- 설계가 비용과 정확도를 결정

실험안 구성
- 대조군
- 하나 이상의 실험군

기능 최적화를 위한 실험
- 하나의 가설, 여러 구현안 비교
    - 한번의 실험, 실험 비용이 증가하나 더 많은 실험을 통해 어느 안이 더 좋을지 빠르게 확인 가능

표준편차가 낮을경우
- 필요 표본 수 감소
- 실험 기간 단축
- 검정력 증가
- MDE 감소


실험 단위는 지표와 함께 지정

실험 대상이란?
- 실험에 포함할 범위, 조건을 만ㅁ족하는 단위의 집합, 표준편차를 최소화 하는 방향으로 정해져야함
