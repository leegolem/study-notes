# 본격적인 가설검정

신약이 혈압에 효과가 있다 를 검정하려 함

A집단: 신약을 투여 -> 평균 μA
B집단: 위약(가짜) 투여 -> 평균μB

통계학적 가설표현
- 귀무가설 H0 신약엔 효과가 없음  μA == μB
- 대립가설 H1 신약엔 효과가 존재 μA != μB

## 짚고 넘어가기: 왜 귀무가설부터 설정할까?
먼저 효과 없음 세계를 가정. <br>
현실 데이터가 그 세계에서 얼마나 드문지 확인
너무 드물면 H0은 맞지 않다 보고 기각

검정은 부정 가정 후 모순 찾기 == 귀류법

생활 비유: 법정 재판
- H0 무죄추정 == 피고인은 무죄
- H1 유죄추정 == 합리적 의심없이 유죄
    - 검사는 무죄 세계에선 나올수 없는 증거 제시가 필요
    - 증거가 애매할경우 무죄 유지 (H0 유지)

생활 비유: 막걸리 맛보기
- H0                          맛이 원래와 같음
- H1 맛이 달라짐
    - 국자 맛이 원래 막걸리에서 나올 수 없을정도로 가벼움
    - H0 기각, H1 채택

> 가설검정이란 H0의 세계를 가정하고,  현실 데이터가 그 세계에서 얼마나 드문지 재는 절차

## p-value?
- 주사위에서 6이 10번 연속으로 -> H0을 의심
- 무죄를 가정했음에도 드문 증거가 나옴 -> H0을 의심
- 클릭률 차이 3% -> H0 의심
- 신공정 불량률이 3배 -> H0 의심



유의수준
- p값: H0이 참일때 현실에서 관찰된 결과 또는 극단적인 결과가 나올 확률
- 유의수준 α: 분석 전에 정해두는 판정 기준성 보통 0.05 H0이 참임에도 5%확률로 틀릴수도 있음

판정법
- p <= α 
    - H0 기각 효과 없음 세계에서 드문 결과 -> 유의미하다
- p > α H0 기각하지 못한다. 
    - -> 효과 없음 세계에서도 흔하게 발생하는 것
    - 주의! H0의 증명이 아닌 기각할 근거가 부족한것


양측 vs 단측
- 양측 검정: 좌우 꼬리 모두를 고려, 효과의 증감을 잡아내고 싶을 때 사용
- 단측 검정: 한쪽만 고려, 증가 혹은 감소만 잡고 싶을 때

4) 3개 이상의 평균을 비교 - 분산분석(anova)

case 1 집단 평균이 비슷 (h0 기각 못함)
- 세 집단 평균이 거의 동일
- 집단 내 변동과 집단 간 변동이 유사
- 결과 F값 작음 -> H0 기각 못함
    - F값? 집단 간 분산 / 집단 내 분산 -> 1에 가까워질수록 집단간 차이가 없음 !중요!

case 2: 집단 평균이 다름 (H0기각)

F값: ANOBA에서 사용하는 검정통계량
- 집단 간 분산 / 집단 내 분산
    - 이 비율이 클수록 집단 간 차이가 크다고 판단

f분포: 두 개의 독립적인 카이제곱 분포의 비율로 만들어지는 분포

ANOVA는 3개 이상 집단의 평균 차이를 한 번에 검증하는 방법, 분산구조를 분석함으로써 평균 차이를 판단

f분포와 기각역

F분포 오른쪽 꼬리 영역이 기각역

관측된 F값이 임계값을 넘으면 p < 0.05 -> H0기각
넘지못하면 H0 기각 못함
