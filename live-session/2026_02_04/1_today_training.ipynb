# 통계 1차시: 기본 개념 이해

# 통계적 사고 중요성

## 통계적 사고가 왜 중요할까?

### 1. 많은 데이터를 한눈에 이해하기 위해서
현실의 데이터는 너무 많기 때문에 하나씩 보는것이 어렵다.<br>
통계는 데이터를 몇개의 숫자로 요약 가능하다
- 평균 주문 금액
- 중앙값
- 분위수

이런 값들만 보더라도
- 고객이 얼마를 사용했는가?
- 일부의 큰 주문이 평균을 왜곡하는가?

를 빠르게 판단 가능해진다.

> 통계는 복잡한 데이터를 사람이 이해가능하도록 형태를 바꾸는 도구다.


### 2. 근거 있는 의사결정을 위해서
통계를 알면 단순 느낌이 아닌 숫자로 판단 가능해진다.
- 표본 수가 충분한지 확인
- A/B 테스트를 통해 두 안을 비교
- p-value나 신뢰구간으로 결과의 불확실성 판단

> 통계는 결정을 정당화하는 근거를 만들어준다.

### 3. 원인과 결과를 구분하기 위해서 (중요한 사회적 질문에 대답하기 위해서)
현실에서는 여로 요인이 동시에 영향을 준다. 통계는 이 요인들을 분리해서 볼 수 있게 만든다
- 회귀 분석
- 공변량 통제
- 준실험 설계

를 통해 "정말로 이 요인에 의해 달라졌는가?"를 따져볼수 있다.

> 통계는 상관관계와 인과관계를 구분하는 데 필수적이다.

### 4. 데이터 속 숨은 패턴을 찾기 위해서
겉으로 보면 보이지 않는 행동 패턴이 데이터에는 숨어 있다.
- 특정 상품을 함께 구매하는 경향
- 특정 시간대, 지역에서 반복되는 행동
- 문제 발생 직전에 나타나는 공통 신호

이러한 패턴들은
- 상관관계
- 조건부 확률
- 군집 분석

과 같은 통계적 방법으로 발견 가능하다.

> 통계적 사고는 감이 아니라 구조로 세상을 보게 만든다.

### 5. 정책이나 서비스의 효과를 평가하기 위해서
어떤 정책이나 서비스가 정말 효과가 있었는지는 비교 없이는 알 수 없다.
시행 전과 후를 비교하고 대상 집단과 비교 집단을 나누어 변화가 우연인지 아닌지 판단할 수 있다.

> 통계는 "이게 실제로 도움이 됐는가?"에 답하는 도구다.

### 6. 잘못된 주장이나 조작을 걸러내기 위해서
통계를 알면 숫자에 속지 않게 된다.
- 편향된 데이터인가? (편향 분석)
- 분포가 왜곡되어 있는가? (분포 비교)
- 이상치로 결과를 부풀린 것은 아닌가? (이상치 탐지)

> 통계는 숫자를 비판적으로 읽는 힘을 길러준다.

정리<br>
통계적 사고란? <br>
데이터를 요약하고, 비교하고, 의심하고, 판단하는 방법이다.

> 데이터 분석에서 통계는 선택이 아닌 기본기이다.

# 데이터의 유형
우리들은 삶을 살면서 여러가지 데이터는 생성되고 받고 있다.
> 데이터 분석과 통계에서 중요한 것은 현실의 정보를 분석 가능한 형태로 구분하는 것이다.

## 1. 수치형 데이터 (Numerical / Quantitative)
숫자로 측정되며, 계산이 가능한 데이터

- 이산형 (Discrete)
    - 셀 수 있는 값
    - 정수 단위로 증가
    - 예: 고객 수, 클릭 횟수, 결함 개수

- 연속형 (Continuous)
    - 측정을 통해 얻는 값
    - 이론적으로 무한히 세분 가능
    - 예: 키, 몸무게, 온도

## 2. 범주형 데이터 (Categorical / Qualitative)
값의 크기보다 구분 자체가 중요한 데이터

- 명목형 (Nominal)
    - 순서가 없음
    - 예: 혈액형, 지역, 제품 카테고리

- 순서형 (Ordinal)
    - 순서는 있지만 간격의 의미는 없음
    - 예: 만족도 등급, 선호도 순위

- 이진형 (Binary)
    - 두 가지 값만 가짐
    - 범주형 데이터의 특수한 형태
    - 예: 0/1, Yes/No

핵심정리
- 수치형 데이터는 계산의 대상이다.
- 범주형 데이터는 분류의 대상이다.
- 데이터 유형에 따라 사용할 수 있는 통계 기법과 분석 방법이 달라진다.

# 통계의 양대산맥?

## 1. 기술통계 (Descriptive Statistics)
이미 가지고 있는 데이터를 요약하고 설명하는 데 목적이 있다.
- 평균, 중앙값 같은 대표값
- 분산, 표준편차 같은 산포도

를 사용해 데이터의 전반적인 특징을 빠르게 파악한다.<br>
단, 요약된 숫자만으로는 데이터의 모든 세부나 이상치를 완전히 설명할 수는 없다.

> 기술통계는 데이터를 보기 쉽게 정리하고 현 상태를 이해하는 데 집중한다.

## 2. 추론통계 (Inferential Statistics)
일부 데이터(표본)를 이용해 전체 데이터(모집단)를 추정하는 것이 목적이다.
- 신뢰구간
- 가설검정

같은 방법을 사용해 관측된 결과가 우연인지, 의미 있는 차이인지를 판단한다.

> 추론통계의 핵심은 "표본으로부터 어디까지 일반화할 수 있는가"이다.

정리
- 기술통계는 데이터를 요약한다.
- 추론통계는 표본으로 전체를 판단한다.
- 분석은 항상 기술통계로 시작해, 필요할 때 추론통계로 넘어간다.

# 통계의 양대산맥: 기술통계
데이터의 상태를 숫자로 요약해 빠르게 파악하는 방법

짚고 넘어가기
1. 평균 (Mean)<br>
모든 값을 더한 뒤, 데이터 개수로 나눈 값.

2. 중앙값 (Median)<br>
데이터를 크기 순으로 정렬했을 때, 가운데에 위치한 값.

3. 분산 (Variance)<br>
각 데이터가 평균에서 얼마나 떨어져 있는지를 제곱해서 평균낸 값. <br>
데이터의 흩어짐 정도를 수치로 나타낸다

4. 표준편차 (Standard Deviation)<br>
분산에 제곱근을 취한 값이다.<br>
분산을 사람이 해석하기 쉬운 형태로 바꾼 것이다.

5. 분산과 표준편차의 관계<br>
둘 다 데이터의 변동성(흩어짐)을 나타낸다<br>
분산은 계산을 위한 값, 표준편차는 해석을 위한 값

## 위치추정
데이터가 어디에 모여 있는지를 한 숫자로 표현하는 방법

### 평균

#### 산술평균 (Arithmetic Mean)
모든 값을 더해 데이터 개수로 나눈 값, 데이터의 전반적인 수준을 나타낸다.
```
평균 = (모든 값의 합) / 데이터 개수
ex) 시험 점수의 평균
```
- 모든 데이터를 사용한다
- 이상치의 영향을 크게 받는다

#### 절사 평균 (Trimmed Mean)
데이터를 정렬한 뒤, 양쪽 끝의 일부 값을 제거하고 계산한 평균이다.<br>
극단값의 영향을 줄이기 위해 사용한다.
- 평균보다 안정적이다
- 중앙값보다는 정보를 더 많이 사용한다

ex) 상·하위 5%를 제거한 평균 소득

#### 가중평균 (Weighted Mean)
각 값에 중요도를 반영해 계산한 평균이다.<br>
모든 값이 동일한 영향력을 갖지 않을 때 사용한다.
```
가중평균 = (값 × 가중치의 합) / 가중치의 합
ex) 과제 40%, 중간 30%, 기말 30% 성적
```

### 중간값 / 중앙값 (Median)
데이터를 정렬했을 때 가운데에 위치한 값<br>
데이터 개수가 짝수인 경우 가운데 두 값의 평균을 사용한다.
- 이상치에 거의 영향을 받지 않는다
- 데이터의 “전형적인 값”을 잘 나타낸다

ex) 소득 분포의 중심값

### 특잇값 / 극단값 (Outlier) == 이상치
다른 값들과 유난히 멀리 떨어진 데이터다. <br>
잘못된 값일 수도 있고 드물지만 의미 있는 값일 수도 있다

극단값이 있으면
- 평균은 크게 변할 수 있다.
- 중앙값은 비교적 안정적이다.

그래서 위치추정 시에는 극단값의 존재 여부를 먼저 확인하는 것이 중요하다.

## 변이추정
데이터가 평균 주변에 얼마나 흩어져 있는지 측정하는 방법

### 평균 기반 변이

#### 편차 (Deviation)
각 데이터가 평균에서 얼마나 떨어져 있는지를 나타낸 값
```
편차 = 각 값 - 평균
```
- 평균보다 크면 양수
- 평균보다 작으면 음수
- 그대로 합하면 +, −가 서로 상쇄된다

#### 분산 (Variance)
- 편차의 상쇄 문제를 해결하기 위해 편차를 제곱해서 평균낸 값이다.
```
분산 = (편차^2들의 평균)
```
- 값이 클수록 데이터가 넓게 퍼져 있다
- 단위가 제곱이라 해석이 직관적이지 않다

> 제곱: + / − 상쇄를 막기 위해

#### 표준편차 (Standard Deviation)
- 분산에 제곱근을 취한 값, 사람이 해석하기 쉽게 만든 변이 지표다.
```
표준편차 = √분산
```
- 원래 데이터와 같은 단위를 가진다
- "보통 평균에서 이 정도 벗어난다"로 해석한다

> 루트: 단위를 되돌려 직관적으로 해석하기 위해

### 정리
- 분산은 계산을 위한 지표다.
- 표준편차는 해석을 위한 지표다.
- 변이를 볼 때는 보통 표준편차를 사용한다.
- 데이터가 고르게 분포되어 있을 때 효과적이다
- 극단값이 있으면 값이 크게 흔들릴 수 있다

간단한 예시

- 두 반의 평균 점수는 같다 → 표준편차가 작은 반이 더 안정적이다
- 두 상품의 평균 매출은 같다 → 표준편차가 큰 상품은 매출 변동이 크다

### 순서 기반 변이
- 데이터의 순서에만 집중한다
- 중앙에 얼마나 몰려 있는지를 본다
- 이상치에 매우 강하다

#### 사분위범위 (Interquartile Range, IQR)
정렬된 데이터에서<br>
25% 지점(Q1)과 75% 지점(Q3) 사이의 거리다.
```
IQR = Q3 − Q1
```
- 가운데 50% 데이터의 퍼짐 정도를 나타낸다
- 극단값의 영향을 거의 받지 않는다

특징
- 분포가 한쪽으로 치우친 경우에 유리하다
- 소득, 매출처럼 이상치가 자연스럽게 존재하는 데이터에 적합하다

ex) 소득 분포, 매출 데이터, 사용자 활동량

### 정리
- 분산과 표준편차는 평균을 기준으로 변이를 본다
- IQR은 순서를 기준으로 변이를 본다
- 데이터에 이상치가 많다면 IQR이 더 안정적이다

## 데이터 분포와 탐색
데이터가 어떤 모양으로 퍼져 있는지 살펴보는 과정

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns

data = [85, 90, 78, 92, 88, 76, 95, 89, 84, 91]
````

### 백분위수와 박스플롯

백분위수?<br>
데이터를 정렬했을 때, 특정 값이 전체에서 어느 위치(하위 몇 %)에 해당하는지를 나타내는 기준.

````python
plt.boxplot(data)
plt.title('boxplot')
plt.show()
````

박스플롯
백분위수를 이용해 데이터 분포를 간단히 시각화한 그래프.
- 상자의 아래: 25% 백분위수(Q1)
- 상자의 위: 75% 백분위수(Q3)
- 상자 안의 선: 중앙값
- 수염(Whisker): 데이터의 범위
- 수염 밖의 점: 이상치 (outliers)
- 맨 끝에 위치한 수평선: 최댓값, 최솟값

상자그림은 중심, 변이(IQR), 이상치를 한 번에 보여준다.<br>
ex) 반별 시험 점수 분포 비교

### 도수분포표와 히스토그램
데이터를 일정한 구간으로 나누고 각 구간에 값이 몇 개 있는지 정리한 표다.
- 구간 크기에 따라 분포의 모습이 달라질 수 있다
- 너무 크면 특징이 사라지고, 너무 작으면 복잡해진다

````python
plt.hist(data, bins=5)
plt.title('histogram')
plt.show()
````

히스토그램 (Histogram)
도수분포표를 막대그래프로 표현한 것이다.
- x축: 값의 구간
- y축: 해당 구간의 빈도
- 막대는 서로 붙어 있다

히스토그램은 분포의 전체적인 형태를 파악하는 데 적합하다.<br>
예: 사용자 구매 금액 분포

## 밀도추정
데이터의 분포를 연속적인 곡선으로 추정해, 값들이 어디에 얼마나 몰려 있는지를 보여주는 방법.

````python
plt.hist(data, bins=5, density=True)
plt.title('density plot (hist-based)')
plt.show()
````

밀도 그림 (Density Plot)
- 히스토그램을 부드러운 곡선으로 표현한 그래프.
- y축은 개수가 아니라 비율(밀도)을 의미한다
- 분포의 모양을 더 직관적으로 볼 수 있다

````python
plt.violinplot(data)
plt.title('violin plot')
plt.show()
````

바이올린 도표 (Violin Plot)<br>
상자그림에 밀도 그림을 결합한 형태다.
- 중앙값과 IQR을 확인할 수 있고 동시에 분포의 모양도 볼 수 있다
- 상자그림보다 데이터가 어떻게 몰려 있는지 더 잘 드러난다.

예: 그룹별 소득 분포 비교

## 이진데이터와 범주 데이터 탐색
범주별로 얼마나 많이 나타나는지를 확인하는 방법
- 이진·범주 데이터는 평균이나 표준편차로 요약하지 않는다.
- 어떤 범주가 얼마나 자주 등장하는지가 핵심이다.

막대도표 (Bar Chart)<br>
각 범주의 빈도나 비율을 막대의 높이로 나타낸 그래프다.
- x축: 범주
- y축: 빈도 또는 비율

히스토그램과의 차이?

- 막대도표: 범주형 데이터, 막대가 떨어져 있음

그래서 실제 분석에서는 막대도표를 더 자주 사용한다.

그래서 실제 분석에서는 막대도표를 더 자주 사용한다.

## 상관관계
데이터들끼리 서로 관련이 있는지 확인하는 방법
- 상관관계는 "함께 움직이는가"를 본다.
- 원인과 결과를 의미하지는 않는다.
- 숫자로는 상관계수, 그림으로는 산점도를 사용한다.

### 상관계수 (Correlation Coefficient)
두 수치형 변수 간의 관계 강도를 숫자로 나타낸 값이다.
- 값의 범위: -1 ~ +1
- +1에 가까울수록: 함께 증가
- -1에 가까울수록: 하나가 증가하면 다른 하나는 감소
- 0에 가까울수록: 뚜렷한 관계 없음

일반적으로 사용하는 것은 피어슨 상관계수다. <br>
ex) 키와 몸무게의 상관관계

### 상관행렬 (Correlation Matrix)
여러 변수들 간의 상관계수를 한 번에 정리한 표다.
- 행과 열: 변수
- 각 칸: 두 변수 간의 상관계수

여러 변수의 관계를 빠르게 훑어볼 때 사용한다.

### 산점도 (Scatter Plot)
두 변수의 관계를 점으로 표현한 그래프다.
- x축: 변수 1
- y축: 변수 2

점들의 패턴으로 관계를 직관적으로 확인한다

상관계수는 숫자, 산점도는 모양을 확인하는 도구다.<br>
ex) 공부 시간과 시험 점수

### 주의할 포인트
상관관계: 두 변수 간의 관계 (원인과 결과가 불분명)<br>
인과관계: 한 변수가 다른 변수에 미치는 영향 (원인과 결과가 분명함)

> 상관관계가 있다고 해서 원인이라고 단정할 수는 없다.

# 통계의 양대산맥: 추론통계

## 추론통계란?
표본을 이용해 모집단의 특성을 추정하는 통계 방법
- 주로 신뢰구간, 가설검정 등을 사용
- 데이터의 일부를 가지고 데이터 전체를 추정하는 것이 핵심

전수조사는 시간·비용·노력의 한계가 있다. 그래서 표본을 뽑아 전체를 추정한다

> 일부 데이터(표본)를 가지고 전체 데이터(모집단)를 판단한다.

- 추론통계의 특징
    - 표본으로 모집단을 추정한다
    - 결과에는 항상 불확실성이 따른다
    - 표본을 어떻게 뽑았는지가 매우 중요하다

- 추론통계에서 주로 하는 일
    - 모수 추정
    - 가설 검정

대표 도구는 신뢰구간과 가설검정이다.

### 잠깐 짚고 넘어가기
1. 신뢰구간 (Confidence Interval)<br>
모집단의 값이 어느 범위 안에 있을지를 추정한 구간<br>
보통 95% 신뢰구간을 사용한다
- 이 방법을 여러 번 반복했을 때, 95%는 이 구간 안에 실제 값이 포함된다는 의미다
- ex: 평균 만족도가 75점, 신뢰구간이 70~80점<br> → 실제 평균은 이 범위 안에 있을 가능성이 높다

2. 가설검정 (Hypothesis Testing)<br>
모집단에 대해 세운 가설이 우연인지, 의미 있는 차이인지를 판단하는 방법이다.
- 귀무가설(H0): 변화나 효과가 없다
- 대립가설(H1): 변화나 효과가 있다
- p-value로 귀무가설을 기각할지 결정한다

예시:
- H0: 새로운 교육 프로그램은 성적에 영향을 주지 않는다
- H1: 새로운 교육 프로그램은 성적에 영향을 준다
